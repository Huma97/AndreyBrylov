## Центральная предельная теорема

Генерируем выборку из экспоненциального распределения ($\lambda = 1$) объемом 1000 и строим по ней гистограмму. Поверх строим теоретическую функция распределения.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

expon_rv = sts.expon(loc = 0, scale = 1)
sample = expon_rv.rvs(1000)

x = np.linspace(0,5,1000)
pdf = expon_rv.pdf(x)
plt.plot(x, pdf)
plt.title('PDF(Exponential)')
plt.ylabel('$f(x)$, fraction of samples')
plt.xlabel('$x$')
plt.hist(sample, normed=True)  
plt.legend()
```

Определяем функция average, которая по выборке вычисляет выборочное среднее. Далеее генерируем 1000 выборок объемом n и строим гистограммы распределений выборочных средних. 

Считаем параметры нормального распределения, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних:
$\mu$ = 1/$\lambda$ = 1, $\sigma$ = $\sqrt{1/n\cdot\lambda ^{2}}$ = 1/$\sqrt{n}$

Поверх каждой гистограммы рисуем плотность соответствующего нормального распределения 

Объем выборки n = 5, $\mu$ = 1, $\sigma$ = 1/$\sqrt{5}$

```python
def average(sample):
    av = 0
    for i in range(len(sample)):
        av += float(sample[i])/len(sample)
    return(av)

samples = []
averages = []

for i in range(1000):
    samples.append(expon_rv.rvs(5))
    averages.append(average(samples[i])) 
norm_rv = sts.norm(loc = 1, scale = (1./5) ** (0.5))
x = np.linspace(0,5,1000)
pdf = norm_rv.pdf(x)
plt.ylabel('$f(x)$, fraction of samples')
plt.xlabel('$x$')
plt.plot(x, pdf)
plt.hist(averages, normed=True)
plt.legend(loc='upper left')
```
